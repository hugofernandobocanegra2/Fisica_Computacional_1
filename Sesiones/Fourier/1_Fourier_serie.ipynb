# Fourier series

For a function $f(x)$ to be represented by a Fourier series expansion, it must satisfy the **Dirichlet conditions**: it must be periodic, single-valued, and continuous (or have a finite number of finite discontinuities). Furthermore, it must have a finite number of maxima/minima within one period and be absolutely integrable.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's see two examples of discret Fourier analysis.

## Example 1: For the Heaviside function:

$$f(x) = 1,\hspace{1cm} 0 <x < \pi$$

$$f(x) = 2,\hspace{1cm}  \pi <x < 2\pi\,.$$

Let's use the basis:  $\{\phi_n(x)\}=\{\dfrac{1}{\sqrt{2L}}e^{i\,n\pi x/L}\}$ of the Hilbert space of function of $\mathscr{L}^2$ with $p(x)=1$ in $c\leq x \leq c+2L$ and $n=(-\infty ... 0 ...+\infty)$.
The Fourier expansion for $f(x)$ is given by:


\begin{equation}
    f(x)= \dfrac{3}{2} -\dfrac{2}{\pi}\sum_{n=0}^{\infty}\dfrac{\sin((2n+1)x)}{2n+1}
\end{equation}

.. see physical mathematica book.

```python
#Programing the f(x) function:
def f(n,x):
    sum = 0 #ini the sum in 0
    for i in range(n):
      sum += np.sin((2*i+1.)*x)/(2*i+1.)

    f = (3./2.) -(2./np.pi)*sum
    return f
```

```python
a = np.linspace(0, 2*np.pi, 100)
plt.figure(figsize=(6,4))
plt.hlines(1,np.pi,0)
plt.hlines(2,2*np.pi,np.pi)
plt.vlines(np.pi,1,2,linestyles='dashed')
plt.plot(a,( 3/2 - 2/np.pi*(np.sin(a)+np.sin(3*a)/3 +np.sin(5*a)/5 +np.sin(7*a)/7) ),'r-',label='')
plt.plot(a, f(4,a), 'b.',label='n=3') #PRogramed function
plt.title('Fourier series of a step function',size=10)
plt.xlabel(r'$x$',size=15)
plt.ylabel(r'$f(x)$',size=12)
plt.legend(prop={'size':10},loc='best')
plt.show()
```

```python
plt.figure(figsize=(6,4))
a = np.linspace(0, 2*np.pi, 100)
plt.hlines(1,np.pi,0)
plt.hlines(2,2*np.pi,np.pi)
plt.vlines(np.pi,1,2,linestyles='dashed')

plt.plot(a, f(1,a), lw=1,label='n=1')
plt.plot(a, f(2,a), lw=1,label='n=2')
plt.plot(a, f(3,a), lw=1,label='n=3')
plt.plot(a, f(4,a), lw=1,label='n=4')
plt.plot(a, f(5,a), lw=1,label='n=5')
plt.plot(a, f(6,a), lw=1,label='n=6')
plt.title('Fourier frecuencies',size=10)
plt.xlabel(r'$x$',size=15)
plt.ylabel(r'$f(x)$',size=12)
plt.legend(prop={'size':10},loc='best')

plt.show()
```

```python
a = np.linspace(0, 2*np.pi, 200)
plt.figure(figsize=(6,4))
plt.hlines(1,np.pi,0)
plt.hlines(2,2*np.pi,np.pi)
plt.vlines(np.pi,1,2,linestyles='dashed')
plt.plot(a, f(40,a), 'r-', label='n=40')
plt.title('Fourier expansion',size=10)
plt.xlabel(r'$x$',size=15)
plt.ylabel(r'$f(x)$',size=12)
plt.legend(prop={'size':10},loc='best')
plt.xlim(0,2*np.pi)

plt.show()
```

## Example 2: Función valor absoluto:

$$f(x) =  -x,\hspace{1 cm}   -\pi \leq x \leq 0 $$
$$f(x) =  +x,\hspace{1 cm}   0 < x \leq \pi\,. $$

$\textbf{Expandir}$ la función $f(x)$ en la base discreta, ortonormal de Fourier: ${\phi_n(x)=\dfrac{1}{\sqrt{2L}}e^{i\,n\pi x/L}}$ en $c\leq x \leq c+2L$, $n=-\infty ... 0 ...+\infty$, del espacio de Hilbert de funciones de cuadrado integrable $\mathscr{L}^2$ con peso $p(x)=1$.

Solution:

$$f(x) = \dfrac{\pi}{2}-\dfrac{2}{\pi}\sum_{k=-\infty, \neq 0}^{+\infty}\dfrac{1}{(2k+1)^2}e^{i(2k+1)x}$$

$$\boxed{f(x) = \dfrac{\pi}{2}-\dfrac{4}{\pi}\sum_{k=0}^{ \infty} \dfrac{\cos((2k+1)x)}{(2k+1)^2}}$$

```python
def xabsolut(n,x):
    ''' Absolut value expansion'''
    sum = np.pi/2

    for i in range(0,n):
      sum += -(4/np.pi)*np.cos((2*i+1)*x)/(2*i+1)**2

    return sum
```

```python
a = np.linspace(-np.pi, np.pi, 100)
plt.figure(figsize=(6,4))
plt.plot(a, np.abs(a),lw=3,label='Numpy')
plt.plot(a, xabsolut(1,a),'g-', lw=1,label='n=1')
plt.plot(a, xabsolut(2,a),'m:', lw=2,label='n=2')
plt.plot(a, xabsolut(4,a),'r-', lw=2,label='n=4')

plt.title('Fourier expansion of the absolute value function',size=10)

plt.xlabel(r'$x$',size=15)
plt.ylabel(r'$f(x)$',size=12)
plt.legend(prop={'size':10},loc='best')
plt.ylim(0,3)
plt.show()
```
